```python
import PyPDF2
from google.colab import files
import io

print("Olá! Eu sou a Joana, psicóloga e tech recruiter. Estou aqui para te conhecer um pouco melhor através do seu perfil.")

# Solicitar o upload do arquivo PDF
uploaded = files.upload()

for fn, data in uploaded.items():
    print(f'Recebi o arquivo "{fn}" com {len(data)} bytes. Obrigada por compartilhar!')

    # Ler o conteúdo do PDF
    try:
        pdf_file = io.BytesIO(data)
        reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(reader.pages)
        text = ""
        for page_num in range(num_pages):
            page = reader.pages[page_num]
            text += page.extract_text()

        print("\nAnalisando seu currículo...")

        # Exemplos de palavras-chave comuns em competências técnicas
        tech_keywords = ['Python', 'Java', 'C++', 'JavaScript', 'SQL', 'AWS', 'Azure', 'Docker', 'Kubernetes', 'Machine Learning', 'Data Science', 'Agile', 'Scrum', 'Linux', 'Git', 'DevOps', 'Frontend', 'Backend', 'Fullstack', 'Cybersecurity', 'Osint', 'Hacking']

        found_skills = [keyword for keyword in tech_keywords if keyword.lower() in text.lower()]

        if found_skills:
            print("\nCom base na leitura, identifiquei as seguintes competências técnicas:")
            for skill in found_skills:
                print(f"- {skill}")
        else:
            print("\nNão identifiquei competências técnicas específicas com base nas palavras-chave comuns que procurei.")
            print("Uma análise mais aprofundada ou uma conversa seriam importantes para explorarmos suas habilidades!")

    except Exception as e:
        print(f"\nOcorreu um erro ao processar o arquivo PDF: {e}")
        print("Pode ser que o arquivo esteja corrompido ou em um formato não compatível com esta análise.")

print("\nObrigada por compartilhar seu currículo! Entrarei em contato se seu perfil for compatível com as vagas que temos.")
```
